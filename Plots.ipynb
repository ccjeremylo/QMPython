```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.xkcd() # some warnings with font

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([])
plt.yticks([])
ax.set_ylim([-30, 10])

data = np.ones(100)
data[70:] -= np.arange(30)

plt.annotate(
    'THE DAY I\n STARTED TAKING\n QUANTUM MECHANICS',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

plt.plot(data)

plt.xlabel('time')
plt.ylabel('my confidence in physics')
plt.show()
```

```python
plt.xkcd()

fig = plt.figure(figsize=[10,7])
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.xticks([])
#plt.yticks([])
ax.set_ylim([-6, 10])

xr = np.arange(-3,3,0.1)
plt.title('Harmonic Oscillator')
plt.plot([0,0],[-6,10],color="blue")
plt.plot([-3,3],[0,0],color="blue")
plt.plot(xr, xr*xr,color="green")
plt.savefig("Harmonic_Oscillator_Potential.pdf")
plt.show()
```

```python
plt.xkcd()

fig = plt.figure(figsize=[10,7])
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.xticks([])
#plt.yticks([])
ax.set_ylim([-6, 10])

xr = np.arange(0,3,0.1)
plt.title('Half Harmonic Oscillator')
plt.plot([0,0],[-6,10],color="blue")
plt.plot([-3,3],[0,0],color="blue")
plt.plot(xr, xr*xr,color="green",lw=4)
plt.plot([0,0],[0,10],color="green",lw=4)
plt.savefig("Half_Harmonic_Oscillator_Potential.pdf")
plt.show()
```
